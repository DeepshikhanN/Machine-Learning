```python
#T-test to check the dependency of smoking on charges
Ho='Charges of smokers and non-smokers are same'#stating null hypothethis
Ha='Charges of smokers and non-smokers are not the same'#stating alternate hypothethis

x=np.array(df[df.smoker=='yes'].charges)#selectng charges corres to smokers as an array
y=np.array(df[df.smoker=='no'].charges)#selectng charges corres to non-smokers as an array
t,p_value=stats.ttest_ind(x,y,axis=0)#performing an Independent T-test
if p_value<0.05:#setting our significance level at 5%
    print(Ha)
else:
    print(Ho)
print(p_value)
```

```python
#T-Test for checking if Gender has an effect on bmi
Ho='Gender has no effect on bmi'#stating null hypothethis
Ha='Gender has effect on bmi'#stating alternate hypothethis

x=np.array(df[df.sex=='male'].bmi)#selectng charges corres to smokers as an array
y=np.array(df[df.sex=='female'].bmi)#selectng charges corres to non-smokers as an array
t,p_value=stats.ttest_ind(x,y,axis=0)#performing an Independent T-test
if p_value<0.05:#setting our significance level at 5%
    print(f'{Ha} as the p_value({p_value.round()})<0.05')
else:
    print(f'{Ho} as the p_value({p_value.round(3)})>0.05')
print(p_value)
```

```python
#Chi square testing to check if smoking habits are different for different genders
Ho='Gender has no effect on smoking habits'
Ha='Gender has effect on smoking habits'
crosstab=pd.crosstab(df['sex'],df['smoker'])#Contingency table of sex and smoker attributes
chi,p_value,dof,expected=stats.chi2_contingency(crosstab)
if p_value<0.05:
    print(f'{Ha} as the p_value({p_value.round()})<0.05')
else:
    print(f'{Ho} as the p_value({p_value.round(3)})>0.05')
crosstab
```

```python
#Chi square testing to check if smoking habits are different for different genders
Ho='Region has no effect on smoking habits'
Ha='Region has effect on smoking habits'
crosstab=pd.crosstab(df['region'],df['smoker'])#Contingency table of sex and smoker attributes
chi,p_value,dof,expected=stats.chi2_contingency(crosstab)
if p_value<0.05:
    print(f'{Ha} as the p_value({p_value.round(3)})<0.05')
else:
    print(f'{Ho} as the p_value({p_value.round(3)})<0.05')
crosstab
```

```python
#Chi square testing to check if bmi of males is different from that of females
Ho='bmi is not significantly different between males and females'
Ha='bmi is significantly different between males and females'
crosstab=pd.crosstab(df['bmi'],df['sex'])
chi,p_value,dof,expected=stats.chi2_contingency(crosstab)
if p_value<0.05:
    print(f'{Ha} as the p_value({p_value.round(3)})<0.05')
else:
    print(f'{Ho} as the p_value({p_value.round(3)})<0.05')
crosstab
```
